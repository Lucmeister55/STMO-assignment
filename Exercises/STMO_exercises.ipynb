#### Exercise: light refraction and Snell's law

What will the relation between the incoming angles be for the two media. In other words, how will the light beam change direction? Solve this problem with the method of Lagrange multipliers.

Let's denote the incident angle of the light beam in the first medium as $\theta_1$ and the incident angle in the second medium as $\theta_2$. We want to find the relation between these two angles.

We can use Snell's law to describe the relationship between the incident angles and the refractive indices of the two media:

$\frac{\sin(\theta_1)}{\sin(\theta_2)} = \frac{n_2}{n_1},$

where $n_1$ and $n_2$ are the refractive indices of the first and second media, respectively.

$T = \frac{d_1}{c_1} + \frac{d_2}{c_2},$

where $d_1$ and $d_2$ are the distances travelled in the first and second media, respectively.

$L(\theta_1, \theta_2, \lambda) = \frac{d_1}{c_1} + \frac{d_2}{c_2} + \lambda\left(\frac{\sin(\theta_1)}{\sin(\theta_2)} - \frac{n_2}{n_1}\right),$

$\frac{\sin(\theta_1)}{\sin(\theta_2)} - \frac{n_2}{n_1} = 0.$

$\frac{d_1}{c_1} - \lambda\frac{1}{\sin(\theta_2)} = 0,$

$\lambda = \frac{d_1}{c_1\sin(\theta_2)}.$

$\frac{\sin(\theta_1)}{\sin(\theta_2)} = 1.$

$\frac{n_2}{n_1} = 1,$

which means the refractive indices of the two media are equal.
